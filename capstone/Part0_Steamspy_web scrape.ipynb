```python
%%time
data_request = dict()
data_request['request'] = 'all'
data_request['page'] = '0'

df = pd.DataFrame()

while True:
    data = steamspypi.download(data_request)
    new_df = pd.DataFrame.from_dict(data, orient='index').reset_index()
    df = pd.concat([df, new_df], ignore_index=True)
    
    if not data:
        break
    
    data_request['page'] = str(int(data_request['page']) + 1)
app_list = df.sort_values(by=['appid']).reset_index(drop=True)
app_list
```
